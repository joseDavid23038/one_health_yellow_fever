```python
import numpy as np
import pandas as pd
```

```python
for ano in range(1994, 2025):
    try:
        df = pd.read_excel(rf"Populações - Brasil/estimativa_populacao_{ano}.xls", skiprows=2)
        
        if ano > 2011:
            xl_file = pd.ExcelFile(rf"Populações - Brasil/estimativa_populacao_{ano}.xls")
            folha = xl_file.sheet_names[1]
            df = xl_file.parse(folha, skiprows=1)
        
        filtro = df[df.columns[0]].isin(['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 
 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 
 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
)
        df = df[filtro]
        
        df.to_pickle(rf"Populações - Brasil/Todas_Regioes/estimativa_populacao_{ano}.pickle")
    
    except Exception as e:
        print(f"Erro no ano {ano}: {e}")
    
    print(ano, end='\r')
```

```python

dicionario_erros = {}
cidades = set({})
for ano in range(1994,2025):
   try:
      df = pd.read_excel(f"Populações - Brasil\Todas_Regioes\estimativa_populacao_{ano}.xlsx")
      
      c_ano = set(df[df.columns[4]].tolist())      
      cidades = cidades.union(c_ano)
   except:
      print(f"Erro:{ano}")
   print(ano,end='\r')
```

```python
cidades = list(cidades)
```

```python
populacao_SE_anos = {}
populacao_SE_anos['Cidade'] = []
populacao_SE_anos['Estado'] = []
for ano in range(1994,2025):
   try:
      df = pd.read_pickle(f"Populações - Brasil\Todas_Regioes\estimativa_populacao_{ano}.pickle")
      populacao_SE_anos[ano] = []
   except:
      populacao_SE_anos[ano] = []
      pass

print('Criação do dicionário concluida. Iniciando preenchiemento')

n = len(cidades)

for i in range(n):
   print(f'{i} de {n}',end='\r')
   populacao_SE_anos['Cidade'].append(cidades[i])
   for ano in range(1994,2025):
      try:
         
         df = pd.read_pickle(f"Populações - Brasil\\Todas_Regioes\\estimativa_populacao_{ano}.pickle")
         UF,COD,MUNIC,NOME,POP = df.columns
         linha = df.index[df[NOME] == cidades[i]].tolist()[0]
         pop = df[POP][linha] 
         populacao_SE_anos[ano].append(pop)         
              
      except:
         try:
            populacao_SE_anos[ano].append(np.nan)  
         except:
            pass
   estado = df[UF][linha]
   populacao_SE_anos['Estado'].append(estado)
```
